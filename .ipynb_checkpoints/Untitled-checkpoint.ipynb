# Price's Law applied to Movie Revenues

Price's law states that square root of the total number of producers will produce at about of a given output. 

This will be an investigation as to wheether Price's law apllies to a single year (2015 and 2010) and to the total number of movies over all the years provided. 





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

movie_data = pd.read_csv('tmdb-movies.csv')


```

```python
np.sqrt(len(movie_data))
```

```python
Total_revenue = movie_data['revenue'].sum()

Total_revenue_adj = movie_data['revenue_adj'].sum
```

Since it appears that revenue figures are not available for all titles. We'll start by removing the titles without data and then sorting the values in descending order.


```python

sorted_revenue = movie_data.sort_values('revenue_adj', ascending=False, )
```

```python
len(sorted_revenue)
```

```python
sorted_revenue 
```

After sorting the data by revenue in descending order there seems to be an issue with the data. 

For Example The Net(1995) is listed as being in the top ten grossing films all time when adjusted for revenue. 

However it appears there was an error when entering the data in. The number should be 110,627,965 (according to IMDB) and not 1,106,279,658. It appears there was an errant '8' added to the figure for some reason.



Also it seems that revenue is not recorded for all films. So for this reason they should be removed from the data set as revenue is the chief factor we will be looking at. 

Just eyeballing it, it looks like the sample size will still be large enough to get a good indication of whether Price's law seems to apply. 


```python
sorted_revenue = sorted_revenue[sorted_revenue.original_title != "The Net"]

```

```python
revenue_no_zeros = sorted_revenue[sorted_revenue.revenue_adj != 0]
```

```python
len(revenue_no_zeros)
```

```python
np.sqrt(len(revenue_no_zeros))
```

```python
all_top_movies_adj = revenue_no_zeros.iloc[:69]['revenue_adj'].sum()
```

```python
rest_movies_adj =revenue_no_zeros['revenue_adj'].sum()
```

```python
all_top_movies_adj / rest_movies_adj
```

```python
np.sqrt(len(revenue_no_zeros)) / len(revenue_no_zeros)
```

```python
labels = ['top 70 movies', 'other 4779']
sizes = [all_top_movies_adj, (rest_movies_adj - all_top_movies_adj)]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')

plt.show()
```

```python
revenue_no_zeros_descending = revenue_no_zeros.sort_values('revenue')

q_1 = revenue_no_zeros_descending.iloc[ : int(len(revenue_no_zeros_descending) * (1/5))]['revenue'].sum()

q_2 = revenue_no_zeros_descending.iloc[int(len(revenue_no_zeros_descending) * (1/5)): int(len(revenue_no_zeros)*(2/5))]['revenue'].sum()

q_3 = revenue_no_zeros_descending.iloc[int(len(revenue_no_zeros_descending) * (2/5)): int(len(revenue_no_zeros_descending) * (3/5))]['revenue'].sum()

q_4 = revenue_no_zeros_descending.iloc[int(len(revenue_no_zeros_descending)*(3/5)): int(len(revenue_no_zeros_descending)*(4/5))]['revenue'].sum()

q_5 = revenue_no_zeros_descending.iloc[int(len(revenue_no_zeros_descending)*(4/5)): ]['revenue'].sum()
```

```python
plt.plot(['20th', '40th', '60th', '80th', '100th'], [q_1, q_2, q_3, q_4, q_5])
plt.ylabel('revenue')
plt.show()
```

Price's law doesn't seem to work here by using all movies with info for adjusted revenue. 
Though we can still see evidence that a small group of movies were disproportiantely productive. 

The squre root of all the movies came out to be about 1.4% of all the moves produced. This 1.4% produced about 10 times its representation in the population. 

That seems like a lot but not quite up to the 50% levels of Price's law

When quintiles for all the movies are accounted for we can see that the top 20 percent of movies seem to be disproportionately productive compared to the other 80 percent. 

```python
unadjusted_revenue_zeros = sorted_revenue[sorted_revenue.revenue != 0]
```

```python
unadjusted_revenue_zeros
```

```python
len(unadjusted_revenue_zeros)
```

```python
unadj_total =unadjusted_revenue_zeros['revenue'].sum()
```

```python
unadj_top = unadjusted_revenue_zeros['revenue'].iloc[:69].sum()
```

```python
unadj_top / unadj_total
```

```python
labels_1 = ["top 70", "other 4779"]
sizes_1 = [unadj_top, (unadj_total - unadj_top)]

fig1, ax1 = plt.subplots()
ax1.pie(sizes_1, labels=labels_1, autopct='%1.1f%%')
ax1.axis('equal')

plt.show()
```

Unadjusted revenue appears to give a similar result as adjusted revenue. This makes sense as the adjusted revenues to dollars in 2010 terms would be based off the the original revenues. So if they were unadjusted it makes sense that everything would decline in roughly equal proprtions and give a similar result. 

Again we see that about 1.4 percent of the overall population is making up about 10 times its represention in the overall population in reveneue. 

There may be holes in the data and its worth checking revenues for a single year where the data might be better. 


```python
revenue_by_year = sorted_revenue[sorted_revenue['release_year'] == 2015]
```

```python
len(revenue_by_year)
```

```python
np.sqrt(len(revenue_by_year))
```

```python
price_check = revenue_by_year.iloc[:24]['revenue'].sum()
```

```python
year_revenue = revenue_by_year['revenue'].sum()
```

```python
price_check / year_revenue
```

```python
labels = 'top 25', 'other 604'
sizes = [price_check, (year_revenue - price_check)]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')

plt.show()
```

```python
revenue_by_year_descending = sorted_revenue.sort_values('revenue')

q_1 = revenue_by_year_descending.iloc[: int((len(revenue_by_year_descending) * (1/5)))]['revenue'].sum()

q_2 = revenue_by_year_descending.iloc[int((len(revenue_by_year_descending) * (1/5))):int((len(revenue_by_year_descending) * (2/5)))]['revenue'].sum()

q_3 = revenue_by_year_descending.iloc[int((len(revenue_by_year_descending)* (2/5))):int((len(revenue_by_year_descending) * (3/5)))]['revenue'].sum()

q_4 = revenue_by_year_descending.iloc[int((len(revenue_by_year_descending) * (3/5))) : int(len(revenue_by_year_descending) * (4/5))]['revenue'].sum()

q_5 = revenue_by_year_descending.iloc[int(len(revenue_by_year_descending) * (4/5)):]['revenue'].sum()
```

```python
plt.plot(["20th", "40th", "60th", "80th", "100th"], [q_1, q_2, q_3, q_4,q_5])
plt.ylabel('revenue')
plt.show()
```

Here we can see how much of the revenue is generated by the moveies at the upper levels of revenue genration vs. the lower levels. 

In contrast to the line graph for all the movies the revenue generated is more lopsided. The 40th to the 60th ppercentile generated almost no revenue while that was not the case when the data for all the movies were looked at.

# Conclusion

Price's law seems to apply to movie revenues if adjusted by year but not when all the movies for all years were taken into account.   

It appears that the provided data has issues with all numbers related to revenues not being provided and potentially not being accurate. 

Also, it may be more helpful to check this by whether or nor a movie had a theater release as some of the entires in the data appear to have been TV releases. Natually revenue figures for TV realeases will be different and possibly more difficult to collect than Movies release revenues. 

While there seem to be issues with the data I think we can conclude that nothing found will prove Price's Law doesn't apply to movie revenues. And we might reasonably epect to see a relatively small group of those movies, whether its the top 20 percent or the top 1 percent, be disproprtionatly productive compared to the rest of the population.
